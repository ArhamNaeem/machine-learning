```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df =pd.read_csv('gas_prices.csv')
```

```python

# plt.plot(df.Year,df.USA,label='USA',marker='.',markersize=10)
# plt.plot(df.Year,df.UK,label='UK',marker='.',markersize=10)
# plt.plot(df.Year,df.Japan,label='JAPAN',marker='.',markersize=10)
# plt.plot(df.Year,df.France, label='FRANCE',marker='.',markersize=10)
# print(df.columns[1:])
for column in df.columns[1:]:
   plt.figure(figsize=(5,3))
   plt.bar(df.Year,df[column], label=column)
    
   plt.title('Gas price over year in USD')
   plt.xlabel('Year')
   plt.ylabel('USD')
   plt.xticks([1990,1993,1996,1999,2002,2005,2008])
   plt.legend()
   plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
   plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fifa_data.csv')
```

```python
df.head()
```

**Histograms**

```python
bins = [40,50,60,70,80,90,100]
plt.hist(df.Overall,bins=bins)
plt.title('Distribuion of players skill level in FIFA 2018')
plt.xlabel('Skill level')
plt.ylabel('Number of players')
plt.xticks(bins)
plt.show()
```

```python
x = df[df['Preferred Foot']=='Left'].count()[0]
y = df[df['Preferred Foot']=='Right'].count()[0]
plt.pie([x,y],labels=['Left foot','Right foot'], colors=['blue','green'],autopct='%.2f%%')
plt.show()
```

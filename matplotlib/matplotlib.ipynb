```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Basic Graph**

```python
x = np.array([0,1,2,3,4])
y = np.array([0,2,4,6,8])
x2 = np.arange(0,4.5,0.5)

plt.figure(figsize=(5,3),dpi=150)

plt.plot(x2[:5],x2[:5]**2,label='x^2')
plt.plot(x2[4:],x2[4:]**2,'b--',label='x^2')
plt.title('Our first graph!',fontdict={"fontname":"Times New Roman", "fontsize":"20"})
plt.xlabel('X-Axis')
plt.ylabel('y-Axis')
plt.xticks([1,2,3,4,5])
# plt.yticks([0,2,4,6,8])
# plt.plot(x,y,label='2x',color='red',marker='.',linestyle='--',linewidth=2,markersize=10,markeredgecolor='blue')
plt.plot(x,y,'r*--',label='2x')
plt.legend()

plt.savefig('plot.png',dpi=200)

plt.show()
```

**Bar charts**

```python
values= [4,3,10]
labels = ['A','B','C']

bar = plt.bar(labels,values)

bar[0].set_hatch('/')
bar[1].set_hatch('O')
bar[2].set_hatch('*')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simple_linear_regression(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    x_diff = x - x_mean
    y_diff = y - y_mean
    
    # slope
    m = np.sum(x_diff * y_diff) / np.sum(y_diff**2)
    
    # y-intercept
    b = y_mean - m * x_mean
    
    return m, b    
```

```python
def predict(m,b,x):
    return m*x + b
```

**Sample Data**

```python
x = np.arange(1, 7)

# Calculate the squares of the numbers in X and store them in y
y = x**2
```

Testing LR

```python
m , b = simple_linear_regression(x,y)
new_x = np.array([6, 7, 8])
predictions = predict(m, b, new_x)
print("Predictions for new x values:", predictions)
```

**Gradient Descent**

```python

def gradient_descent(x, y, alpha=0.01, iterations=1000):
    n = len(x)
    theta = np.zeros(2)  # Initialize coefficients [alpha, beta] to zeros
    for _ in range(iterations):
        y_pred = theta[0] + theta[1] * x
        error = y_pred - y
        gradient = np.array([np.sum(error), np.sum(error * x)])
        print(theta)
        theta = theta - alpha * (1/n) * gradient
    return theta[0], theta[1]
```

```python
b , m = gradient_descent(x,y,alpha=0.01,iterations=1000)
new_x = np.array([6, 7, 8])
predictions = predict(m, b, new_x)
print("Predictions for new x values:", predictions)
```

Plotting

```python
plt.scatter(x, y, label='Data Points', color='blue')
plt.plot(x, b + m * x, label='Regression Line', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
